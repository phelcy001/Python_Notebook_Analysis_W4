```
#import pandas library
import pandas as pd
#import numpy library
import numpy as np
```

```
#loading the dataset
df = pd.read_csv("http://bit.ly/autolib_dataset")
```

```
#previewing the first 10 rows of the dataset
df.head(10)
```

```
#getting to know about the information 
df.info()
```

```
#checking for missing values across each column
df.isnull().sum()
```

```
#dropping columns with nan values
df1 = df.dropna(axis=1)
df1.head(10)
```

```
#checking for missing values
df1.isnull().values.any()
```

```
#removing duplicates
df1.drop_duplicates().head(10)
```

```
#combining the year, month, day, hour, minute columns and converting them to datetime formats
cols = pd.to_datetime(df1.iloc[:, 18:24])
df1['date'] = cols
df1.head(5)
```

```
#we are going to calculate the difference of the bluecars to determine whether the cars were picked or returned and sort it by the ID and the date
df1['diff_bc'] = df1.groupby(['ID'])['Bluecar counter'].diff().fillna(0)
df1.sort_values(by=['ID','date']).head(10)
```

```
#we find the most popular hour for picking up a bluecar in the city of paris in April 2018
df3 = df1.groupby(['hour', 'City'])['diff_bc'].sum().sort_values(ascending=False).head(1)
df3
```

```
#most popular hour for returning cars
df1.groupby(['hour'])['diff_bc'].sum().sort_values(ascending=False).head(1)
```

```
#most popular hour for returning cars in paris
df1.groupby(['hour','City'])['diff_bc'].sum().sort_values(ascending=False).head(1)
```

```
#most popular station
df1['ID'].mode()
```

```
#most popular station in paris
list = ['Paris']
df8 = df1[df1.City.isin(list)]
df8['ID'].mode()
```

```
#most popular station at the most popular picking hour 
list = [17]
df4 = df1[df1.hour.isin(list)]
df4.groupby('ID')['hour'].count().sort_values(ascending=False).head(1)
```

```
#most popular station at the most popular picking hour in paris
list = [17]
df4 = df1[df1.hour.isin(list)]
df4.groupby(['ID'],['City'] == 'Paris')['hour'].count().sort_values(ascending=False).head(1)
```

```
#postal code most popular for picking bluecars
df1.groupby(['Postal code'])['diff_bc'].sum().sort_values(ascending=True).head(1)
```

```
#postal code most popular for picking bluecars in paris
list = [17]
df4 = df1[df1.hour.isin(list)]
df4.groupby(['Postal code'])['diff_bc'].sum().sort_values(ascending=True).head(1)
```
